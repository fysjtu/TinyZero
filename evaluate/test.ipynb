```python
import pandas as pd  # pandas 提供 read_parquet 等数据处理函数
from pathlib import Path  # 统一处理本地/相对路径
# 设置显示最大列宽为“无限制”
pd.set_option('display.max_colwidth', None)

# 如果列数太多也被折叠了，可以加这一行
pd.set_option('display.max_columns', None)

# 1. 指定 parquet 文件路径，可替换成你实际的训练/验证数据
parquet_path = Path("/root/TinyZero/data/test.parquet")

# 2. 读取 parquet 文件
#    pandas 会自动选择可用的引擎（pyarrow/fastparquet），若缺少依赖需要先 pip 安装
df = pd.read_parquet(parquet_path)

# 3. 了解整体情况
print(f"总行数：{len(df)}, 列：{df.columns.tolist()}")

# 4. 展示前几行，核对字段是否正确
print(df.head(5))  # 如果在 notebook 中；在脚本里可用 print(df.head(5))
```
